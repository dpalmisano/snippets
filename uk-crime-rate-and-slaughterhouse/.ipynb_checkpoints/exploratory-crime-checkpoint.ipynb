```python
import pandas as pd
```

```python
def crime_counts_by_msoa():
    crime = pd.read_csv('samples/2017-08-sussex-street.csv', usecols=[1, 7,8,9])
    crime['msoa'] = crime['LSOA name'].str[:-1]
    return crime.groupby(['msoa']).size().reset_index(name='counts')
```

```python
residents = pd.read_csv('samples/census-msoa-residents-sample.csv')
```

```python
residents
```

```python
crime = crime_counts_by_msoa()
```

```python
crime_and_residents = pd.merge(residents, crime, on='msoa', how='inner')
# crime_and_residents
crime_and_residents['ratio'] = crime_and_residents.apply(lambda row: row.counts / row.residents, axis=1)
crime_and_residents
```
